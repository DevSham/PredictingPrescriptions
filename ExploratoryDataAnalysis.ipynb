```python
# Phase 4: Finding Relationships between the variables or Finding patterns
```

```python
# Check how products and their alternatives are distributed among diffent diagnoses.
```

```python
pivot_table = data_with_required_fields_and_no_missing_values.pivot_table(
    index=['age unit', 'Age Range', 'Diagnosis'],
    values=['ALTERNATIVE PRODUCT 1', 'ALTERNATIVE PRODUCT 2', 'PRODUCT DESCRIPTION/ BRAND'], 
    aggfunc='count',
    fill_value=0
)
```

```python
# Flatten the pivot table
flat_table = pivot_table.reset_index()
```

```python
# Combine the counts of alternatives into a single category for easier plotting
flat_table['Combined Alternatives'] = (
    flat_table['ALTERNATIVE PRODUCT 1'] + 
    flat_table['ALTERNATIVE PRODUCT 2'] + 
    flat_table['PRODUCT DESCRIPTION/ BRAND']
)
```

```python
norm = plt.Normalize(vmin=flat_table['Combined Alternatives'].min(), vmax=flat_table['Combined Alternatives'].max())
colors = plt.cm.viridis(norm(flat_table['Combined Alternatives']))
```

```python
plt.figure(figsize=(14, 10))

# Create scatter plot
scatter = plt.scatter(
    flat_table['Age Range'],  # x-axis
    flat_table['Diagnosis'],  # y-axis
    s=flat_table['Combined Alternatives'] * 10,  # Size of the dots
    c=colors,  # Color based on combined alternatives
    alpha=0.7,
    edgecolors='w',
    linewidth=0.5
)

# Adding labels and title
plt.xlabel('Age Range')
plt.ylabel('Diagnosis')
plt.title('Scatter Plot of Age Range, Age Unit, and Diagnosis with Alternative Products')
plt.xticks(rotation=45)  # Rotate x-axis
plt.grid(True)

# Create a colorbar with the ScalarMappable
cbar = plt.colorbar(plt.cm.ScalarMappable(norm=norm, cmap='viridis'), ax=plt.gca(), label='Combined Alternatives Count')

plt.tight_layout()
plt.show()
```

```python
# Plotting heatmap to find the number of products and alternatives per diagnosis.
plt.figure(figsize=(18, 8))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu')
plt.title('Heatmap of Products by Age Range, Age Unit, and Diagnosis')
plt.xlabel('Products and Alternatives')
plt.ylabel('Diagnosis, Age Range, Age Unit')
plt.show()
```
