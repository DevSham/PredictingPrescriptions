```python
# Phase 3 Cleaning the data
```

```python
# Dropping none required Fields
```

```python
data_with_required_fields = data.drop(['APPROPRIATE ADD ON', 'Comments', 'Age Range 1', 'Age Range 1', 'Age Range 2', 'Contraindications', 'Unnamed: 12', 'Unnamed: 13', 'Age Range.1'], axis=1)
```

```python
# Confirming if none required fields have been remove.
data_with_required_fields.columns
```

```python
# Looking for missing values.
data_with_required_fields.isnull().sum()
```

```python
# Getting rid of records with missing values.
data_with_required_fields_and_no_missing_values = data_with_required_fields.dropna(subset=['Diagnosis',  'Age Range', 'age unit', 'PRODUCT DESCRIPTION/ BRAND', 'ALTERNATIVE PRODUCT 1', 'ALTERNATIVE PRODUCT 2']) 
```

```python
# Confirming if missing values have been remove.
data_with_required_fields_and_no_missing_values.isnull().sum()
```

```python
# Check for duplicates
duplicates = data_with_required_fields_and_no_missing_values.duplicated().sum()
```

```python
duplicates
```

```python
# Finding Relationships between the variables or Finding patterns
```

```python
# Summarize data for visualization.
pivot_table = data_with_required_fields_and_no_missing_values.pivot_table(index=['age unit', 'Age Range', 'Diagnosis'],
                            columns=['Diagnosis'],
                            values=['ALTERNATIVE PRODUCT 1', 'ALTERNATIVE PRODUCT 2', 'PRODUCT DESCRIPTION/ BRAND'], 
                            aggfunc='count',
                            fill_value=0)
```

```python
pivot_table
```

```python
# Plotting heatmap
plt.figure(figsize=(18, 8))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu')
plt.title('Heatmap of Products by Age Range, Age Unit, and Diagnosis')
plt.xlabel('Products and Alternatives')
plt.ylabel('Diagnosis, Age Range, Age Unit')
plt.show()
```
